Завдання 2: Створіть двовимірний масив (матрицю) розміром 3x3, заповніть його добутками квадратів кожного числа та його подвійного значення, починаючи з 1, та виведіть його значення.

```python
import numpy as np

rows, cols = 3, 3

matrix = np.zeros((rows, cols))

value = 1
for i in range(rows):
    for j in range(cols):
        matrix[i][j] = (value ** 2) * (2 * value)
        value += 1

print(matrix)
```

```python
import numpy as np

rows, cols = 4, 4

low = 0
high = 1

matrix = np.random.uniform(low, high, (rows, cols))

mean_vector = np.mean(matrix, axis=0)

print("Матриця:")
print(matrix)
print("\nВектор-рядок з середніми значеннями стовпців:")
print(mean_vector)
```

Завдання 5: Створіть два одновимірних масиви розміром 7, заповніть їх випадковими дійсними числами від -2.5 до 2.5 та виконайте на них поелементні операції додавання, віднімання та множення.

```python
import numpy as np

size = 7

low = -2.5
high = 2.5

array1 = np.random.uniform(low, high, size)
array2 = np.random.uniform(low, high, size)

addition_result = array1 + array2
subtraction_result = array1 - array2
multiplication_result = array1 * array2

print("Масив 1:", array1)
print("Масив 2:", array2)
print("Результат додавання:", addition_result)
print("Результат віднімання:", subtraction_result)
print("Результат множення:", multiplication_result)
```

Завдання 8: Створіть матрицю розміром 2x2, заповніть її випадковими цілими числами в діапазоні від 1 до 12 та знайдіть її обернену матрицю

```python
import numpy as np

rows, cols = 2, 2

low = 1
high = 12

matrix = np.random.randint(low, high + 1, (rows, cols))

print("Початкова матриця:")
print(matrix)

try:
    inverse_matrix = np.linalg.inv(matrix)
    print("\nОбернена матриця:")
    print(inverse_matrix)
except np.linalg.LinAlgError:
    print("\nОбернена матриця не існує, оскільки детермінант дорівнює 0.")
```

Завдання 10: Створіть матрицю розміром 3x4 та вектор розміром 4, заповніть матрицю випадковими цілими числами в діапазоні від 1 до 10, а вектор - випадковими дійсними числами від 0 до 1, і знайдіть суму додатних елементів отриманої матриці.

```python
import numpy as np

matrix_rows, matrix_cols = 3, 4
vector_size = 4

low_matrix = 1
high_matrix = 10

low_vector = 0
high_vector = 1

matrix = np.random.randint(low_matrix, high_matrix + 1, (matrix_rows, matrix_cols))

vector = np.random.uniform(low_vector, high_vector, vector_size)

print("Матриця:")
print(matrix)
print("\nВектор:")
print(vector)

positive_elements_sum = np.sum(matrix[matrix > 0])

print("\nСума додатних елементів матриці:", positive_elements_sum)
```

Завдання 15: Створіть матрицю розміром 9x9, заповніть її випадковими дійсними числами в діапазоні від -0.5 до 0.5 та знайдіть вектор-стовпчик, що містить найбільше за модулем значення елементу у кожному рядку матриці.

```python
import numpy as np

rows, cols = 9, 9

low = -0.5
high = 0.5

matrix = np.random.uniform(low, high, (rows, cols))

max_abs_values_vector = np.argmax(np.abs(matrix), axis=1)

print("Матриця:")
print(matrix)
print("\nВектор-стовпчик з найбільшими за модулем значеннями:")
print(max_abs_values_vector)
```
